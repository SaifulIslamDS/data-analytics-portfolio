# Marketing Campaign Performance Analysis Using Python

## Project Objective

Analyze marketing campaign performance using Python to identify which campaign channels, campaign types, customer segments, and target audiences generate the best ROI, engagement, conversion performance, and cost efficiency.

## Tools

- Python
- Pandas
- NumPy
- Matplotlib
- Jupyter Notebook

```python
# 1. Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:,.2f}".format)
```

```python
# 2. Load dataset

raw_path = "../data/marketing_campaign_dataset.csv"

df = pd.read_csv(raw_path)

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
df.head()
```

```python
# 3. Initial data inspection

display(df.info())
display(df.describe(include="all").T)

missing_values = df.isna().sum().sort_values(ascending=False)
duplicates = df.duplicated().sum()

print("Missing values:")
print(missing_values)
print("\nDuplicate rows:", duplicates)
```

```python
# 4. Clean column names

df_clean = df.copy()

df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_", regex=False)
)

df_clean.columns.tolist()
```

```python
# 5. Clean data types and create analysis-ready fields

df_clean["date"] = pd.to_datetime(df_clean["date"], errors="coerce")

df_clean["acquisition_cost"] = (
    df_clean["acquisition_cost"]
    .astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

df_clean["duration_days"] = (
    df_clean["duration"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype(int)
)

df_clean["year"] = df_clean["date"].dt.year
df_clean["month"] = df_clean["date"].dt.month
df_clean["month_name"] = df_clean["date"].dt.strftime("%b")
df_clean["year_month"] = df_clean["date"].dt.to_period("M").astype(str)
df_clean["quarter"] = df_clean["date"].dt.quarter

# Derived marketing KPIs
df_clean["ctr"] = df_clean["clicks"] / df_clean["impressions"]
df_clean["estimated_conversions"] = df_clean["clicks"] * df_clean["conversion_rate"]
df_clean["estimated_return"] = df_clean["acquisition_cost"] * df_clean["roi"]
df_clean["cost_per_click"] = df_clean["acquisition_cost"] / df_clean["clicks"]
df_clean["cost_per_estimated_conversion"] = df_clean["acquisition_cost"] / df_clean["estimated_conversions"]

df_clean.head()
```

```python
# 6. Validate cleaned dataset

quality_summary = {
    "rows": df_clean.shape[0],
    "columns": df_clean.shape[1],
    "missing_values": int(df_clean.isna().sum().sum()),
    "duplicate_rows": int(df_clean.duplicated().sum()),
    "min_date": df_clean["date"].min(),
    "max_date": df_clean["date"].max(),
}

quality_summary
```

```python
# 7. Overall KPI summary

total_campaigns = df_clean["campaign_id"].nunique()
total_cost = df_clean["acquisition_cost"].sum()
avg_cost = df_clean["acquisition_cost"].mean()
avg_roi = df_clean["roi"].mean()
avg_conversion_rate = df_clean["conversion_rate"].mean()
total_clicks = df_clean["clicks"].sum()
total_impressions = df_clean["impressions"].sum()
overall_ctr = total_clicks / total_impressions
avg_engagement = df_clean["engagement_score"].mean()
estimated_conversions = df_clean["estimated_conversions"].sum()
estimated_return = df_clean["estimated_return"].sum()
avg_cpc = total_cost / total_clicks
avg_cpec = total_cost / estimated_conversions

kpi_summary = pd.DataFrame({
    "KPI": [
        "Total Campaigns",
        "Total Acquisition Cost",
        "Average Acquisition Cost",
        "Average ROI",
        "Average Conversion Rate",
        "Total Clicks",
        "Total Impressions",
        "Overall CTR",
        "Average Engagement Score",
        "Estimated Conversions",
        "Estimated Return",
        "Average Cost per Click",
        "Average Cost per Estimated Conversion"
    ],
    "Value": [
        total_campaigns,
        total_cost,
        avg_cost,
        avg_roi,
        avg_conversion_rate,
        total_clicks,
        total_impressions,
        overall_ctr,
        avg_engagement,
        estimated_conversions,
        estimated_return,
        avg_cpc,
        avg_cpec
    ]
})

kpi_summary
```

```python
# 8. Helper function for grouped marketing performance

def campaign_performance(group_col):
    result = (
        df_clean
        .groupby(group_col)
        .agg(
            campaigns=("campaign_id", "count"),
            total_cost=("acquisition_cost", "sum"),
            avg_roi=("roi", "mean"),
            avg_conversion_rate=("conversion_rate", "mean"),
            total_clicks=("clicks", "sum"),
            total_impressions=("impressions", "sum"),
            avg_engagement=("engagement_score", "mean"),
            estimated_conversions=("estimated_conversions", "sum"),
            estimated_return=("estimated_return", "sum")
        )
        .reset_index()
    )

    result["ctr"] = result["total_clicks"] / result["total_impressions"]
    result["cost_per_estimated_conversion"] = (
        result["total_cost"] / result["estimated_conversions"]
    )

    return result.sort_values("avg_roi", ascending=False)

channel_performance = campaign_performance("channel_used")
campaign_type_performance = campaign_performance("campaign_type")
segment_performance = campaign_performance("customer_segment")
audience_performance = campaign_performance("target_audience")

channel_performance
```

```python
# 9. Visualization: average ROI by channel

channel_roi = channel_performance.sort_values("avg_roi", ascending=True)

plt.figure(figsize=(10, 6))
plt.barh(channel_roi["channel_used"], channel_roi["avg_roi"])
plt.title("Average ROI by Marketing Channel")
plt.xlabel("Average ROI")
plt.ylabel("Channel")
plt.tight_layout()
plt.show()
```

```python
# 10. Visualization: cost per estimated conversion by channel

channel_cpec = channel_performance.sort_values("cost_per_estimated_conversion", ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(channel_cpec["channel_used"], channel_cpec["cost_per_estimated_conversion"])
plt.title("Cost per Estimated Conversion by Channel")
plt.xlabel("Cost per Estimated Conversion")
plt.ylabel("Channel")
plt.tight_layout()
plt.show()
```

```python
# 11. Visualization: monthly acquisition cost and estimated return

monthly_performance = (
    df_clean
    .groupby("year_month")
    .agg(
        total_cost=("acquisition_cost", "sum"),
        estimated_return=("estimated_return", "sum"),
        avg_roi=("roi", "mean"),
        campaigns=("campaign_id", "count")
    )
    .reset_index()
)

plt.figure(figsize=(12, 6))
plt.plot(monthly_performance["year_month"], monthly_performance["total_cost"], marker="o", label="Total cost")
plt.plot(monthly_performance["year_month"], monthly_performance["estimated_return"], marker="o", label="Estimated return")
plt.title("Monthly Acquisition Cost vs Estimated Return")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 12. Visualization: conversion rate by campaign type

campaign_type_sorted = campaign_type_performance.sort_values("avg_conversion_rate", ascending=True)

plt.figure(figsize=(10, 6))
plt.barh(campaign_type_sorted["campaign_type"], campaign_type_sorted["avg_conversion_rate"])
plt.title("Average Conversion Rate by Campaign Type")
plt.xlabel("Average Conversion Rate")
plt.ylabel("Campaign Type")
plt.tight_layout()
plt.show()
```

```python
# 13. Top performing campaigns by ROI

top_campaigns_by_roi = (
    df_clean
    .sort_values("roi", ascending=False)
    .head(10)
    [[
        "campaign_id", "company", "campaign_type", "channel_used",
        "target_audience", "customer_segment", "acquisition_cost",
        "conversion_rate", "roi", "clicks", "impressions", "engagement_score"
    ]]
)

top_campaigns_by_roi
```

```python
# 14. Underperforming campaigns

underperforming_campaigns = (
    df_clean
    .query("roi < roi.quantile(0.25) and conversion_rate < conversion_rate.quantile(0.25)")
    .sort_values(["roi", "conversion_rate"], ascending=True)
    .head(20)
)

underperforming_campaigns[
    [
        "campaign_id", "company", "campaign_type", "channel_used",
        "target_audience", "customer_segment", "acquisition_cost",
        "conversion_rate", "roi", "engagement_score"
    ]
]
```

```python
# 15. Save cleaned dataset

output_path = "../cleaned-data/marketing_campaign_cleaned.csv"
df_clean.to_csv(output_path, index=False)

print(f"Cleaned data saved to: {output_path}")
```
